```python
import pandas as pd
from textblob import TextBlob
```

```python
df = pd.read_csv("IDM_dataset.csv")
```

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "positive"
    elif polarity < 0:
        return "negative"
    else:
        return "neutral"

df["predicted_sentiment"] = df["review"].apply(get_sentiment)
```

```python
print(df.head())
```

```python
df['sentiment'] = df['sentiment'].str.strip()
df['predicted_sentiment'] = df['predicted_sentiment'].str.strip()
```

```python
df.to_csv("sentiment_results_fixed.csv", index=False, quoting=1)
```
